### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresoTotal1 = tienda['Precio'].sum()
ingresoTotal2 = tienda2['Precio'].sum()
ingresoTotal3 = tienda3['Precio'].sum()
ingresoTotal4 = tienda4['Precio'].sum()
print(f'El ingreso de la tienda 1 es {ingresoTotal1}')
print(f'El ingreso de la tienda 2 es {ingresoTotal2}')
print(f'El ingreso de la tienda 3 es {ingresoTotal3}')
print(f'El ingreso de la tienda 4 es {ingresoTotal4}')
```

```python
maximo = max(ingresoTotal1,ingresoTotal2,ingresoTotal3,ingresoTotal4)
print(f'El ingreso máximo es {maximo}')
```

# 3. Calificación promedio de la tienda


```python
prom1 = tienda['Calificación'].mean()
prom2 = tienda2['Calificación'].mean()
prom3 = tienda3['Calificación'].mean()
prom4 = tienda4['Calificación'].mean()
summary = {'tienda1':prom1, 'tienda2':prom2, 'tienda3':prom3,'tienda4':prom4}
print(f'El promedio de calificación de la tienda 1 es {prom1}')
print(f'El promedio de calificación de la tienda 2 es {prom2}')
print(f'El promedio de calificación de la tienda 3 es {prom3}')
print(f'El promedio de calificación de la tienda 4 es {prom4}')
```

```python
ordenado = dict(sorted(summary.items(), key=lambda item: item[1],reverse=True))
for key, value in ordenado.items():
    print(f'{key}: {value}')
```
